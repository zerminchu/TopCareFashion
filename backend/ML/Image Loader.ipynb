```python
db = firestore.client()
items_ref = db.collection('Item')
items = items_ref.stream()

Image_urls = []
labels = []

for item in items:
    item_data = item.to_dict()
    category = item_data.get('category', 'Unknown')  # Default to 'Unknown' if no category
    urls = item_data.get('image_urls', [])  # Get the array of image URLs
    for url in urls:
        if url:  # Check if the URL is not None
            Image_urls.append(url)
            labels.append(category)  # Associate each URL with the category
            
#     labels.append(category)
#     Image_urls.append(image_urls)
    
    
# if image_urls and label:  # Check if both image_url and label are not None
#         image_urls.append(image_urls)
#         labels.append(label)
```

```python


def download_image(url, folder=folder_name):
    if not url:
        return

    # Extract filename from URL
    filename = url.split('?')[0].split('/')[-1]
    
    # Ensure filename ends with .jpg
     # Ensure filename ends with .jpg or .png
    if not (filename.lower().endswith('.jpg') or filename.lower().endswith('.png') or filename.lower().endswith('.jpeg')):
        filename += '.jpg'  # Default to .jpg if no extension is present
    
    # Create folder if it doesn't exist
    if not os.path.exists(folder):
        os.makedirs(folder)

    # Full path for the image to be saved
    file_path = os.path.join(folder, filename)

    # Download and save the image
    response = requests.get(url)
    if response.status_code == 200:
        with open(file_path, 'wb') as f:
            f.write(response.content)
        print(f"Image saved to {file_path}")
    else:
        print(f"Failed to download {url}")

# Assuming image_urls is your list of URLs
for url in Image_urls:
    download_image(url)
```

```python
data_for_csv = []
folder_name = 'downloaded_images'
for url, label in zip(Image_urls, labels):
    # Extract filename from URL
    filename = url.split('?')[0].split('/')[-1]
    
    # Ensure filename ends with .jpg or .png
    if not (filename.lower().endswith('.jpg') or filename.lower().endswith('.png') or filename.lower().endswith('.jpeg')):
        filename += '.jpg'  # Default to .jpg if no extension is present

    # Append the folder name
    full_filename = os.path.join(folder_name, filename)

    data_for_csv.append({
        'image': full_filename,
        'label': label
    })
```

```python
csv_file_path = r'C:\Users\Joshua\Machine Learning\FYP\image_labels.csv'
# Writing data to CSV
with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['image', 'label'])
    writer.writeheader()
    writer.writerows(data_for_csv)

print(f"CSV file created: {csv_file_path}")
```
